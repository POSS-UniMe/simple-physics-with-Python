```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(1, (12,8))
ax = fig.gca(projection='3d')

# Preparing a set of points belonging to the xy plane
xx = np.linspace(0, 2, 100)
yy = np.linspace(-2, 2, 100)

xv, yv = np.meshgrid(xx, yy)
zv = xv * 0
#

# Plotting the plane ( the alpha parameter, ranging from 0 to 1, controls opacity)
ax.plot_surface(xv, yv, zv, alpha = 0.05)


# Setting the axes properties
ax.set(xlim3d=(0, 2), xlabel=r'$X$')
ax.set(ylim3d=(-1, 1), ylabel=r'$Y$')
ax.set(zlim3d=(-2, 2), zlabel=r'$Z$')

# Showing a semicircle having the center in O and radius 1
angle = np.linspace(-np.pi/2, np.pi/2, 36)
x = np.cos(angle)
y = np.sin(angle)
ax.plot(x, y, zs=0, zdir='z', label='curve in (x, y)',
        c = 'red')
#

# Showing the velocity vector and its components (tangential and radial)
a = np.array([1, 1, 0])
b = np.array([1, 0, 0])
c = np.array([0, 1, 0])

ax.quiver(1, 0, 0, *a, length=1, color = 'green', linewidth = 4)
ax.quiver(1, 0, 0, *b, length=1, color = 'cyan', linewidth = 4, zorder = -1, alpha = 0.6)
ax.quiver(1, 0, 0, *c, length=1, color = 'lime', linewidth = 4, zorder = -1, alpha = 0.6)


ax.scatter(0, 0, 0, color = 'r', label='O', s = 9)
ax.scatter(1, 0, 0, color = 'b', label = 'P', s = 25)

xline1=np.array([1, 2])
yline1 = np.array([1, 1])
ax.plot(xline1, yline1, zs=0, zdir='z', linestyle = 'dotted')

xline2 = np.array([2, 2])
yline2 = np.array([0, 1])
ax.plot(xline2, yline2, zs=0, zdir='z', linestyle = 'dotted')

# Showing the radial axis 
xline3=np.array([0, 2])
yline3 = np.array([0, 0])
ax.plot(xline3, yline3, zs=0, zdir='z',
        linestyle = 'dashed', c = 'black', zorder = -1)

# Showing the tangential axis
xline4=np.array([1, 1])
yline4 = np.array([-2, 2])
ax.plot(xline4, yline4, zs=0, zdir='z',
        linestyle = 'dashed', c = 'black', zorder = -1)

# Showing an arc of a circle, with center in P, to represent the angle theta
theta = np.linspace(0, np.pi/4, 36)
xt = 1 + 0.4*np.cos(theta)
yt = 0.4*np.sin(theta)

ax.plot(xt, yt, zs=0, zdir='z', label='curve in (x, y)',
        c = 'magenta', linewidth = 5)
ax.text(1.5, 0, 0.1, r"$\theta$", "x", size = 16)
#

# Showing the position of the observer O 
ax.text(-0.1, 0, 0.1, "O", "x", size = 16)

# Showing the position of the particle P
ax.text(0.9, 0, 0.1, "P", "x", size = 16)

ax.grid(False)
plt.show()
fig.savefig('AngleTheta.png')
```
