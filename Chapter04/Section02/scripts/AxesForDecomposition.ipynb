```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(1, (12,8))
ax = fig.gca(projection='3d')

xx = np.linspace(-2, 2, 100)
yy = np.linspace(-2, 2, 100)

xv, yv = np.meshgrid(xx, yy)
zv = xv * 0

ax.plot_surface(xv, yv, zv, alpha = 0.05)

#ax.set_xlim3d(-2, 2)
#ax.set_ylim3d(-2,2)
#ax.set_zlim3d(-2, 2)

# Setting the axes properties
ax.set(xlim3d=(-2, 2), xlabel=r'$X$')
ax.set(ylim3d=(-2, 2), ylabel=r'$Y$')
ax.set(zlim3d=(-2, 2), zlabel=r'$Z$')

# Plot a sin curve using the x and y axes.
#x = np.linspace(0, 1, 100)
#y = np.sin(x * 2 * np.pi) / 2 + 0.5
#ax.plot(x, y, zs=0, zdir='z', label='curve in (x, y)')

angle = np.linspace(0, 2*np.pi, 36)
x = np.cos(angle)
y = np.sin(angle)

ax.plot(x, y, zs=0, zdir='z', label='curve in (x, y)', linestyle = 'dotted')

a = np.array([1, 1, 0])
b = np.array([1, 0, 0])
c = np.array([0, 1, 0])
#m = np.cross(a, b)

ax.quiver(1, 0, 0, *a, length=1, color = 'green', linewidth = 4)
ax.quiver(1, 0, 0, *b, length=1, color = 'cyan', linewidth = 4)
ax.quiver(1, 0, 0, *c, length=1, color = 'lime', linewidth = 4)


ax.scatter(0, 0, 0, color = 'r', label='O', s = 9)
ax.scatter(1, 0, 0, color = 'b', label = 'P', s = 25)

xline1=np.array([1, 2])
yline1 = np.array([1, 1])
ax.plot(xline1, yline1, zs=0, zdir='z', linestyle = 'dotted')

xline2 = np.array([2, 2])
yline2 = np.array([0, 1])
ax.plot(xline2, yline2, zs=0, zdir='z', linestyle = 'dotted')

xline3=np.array([0, 0])
yline3 = np.array([0, 2])
ax.plot(xline3, yline3, zs=0, zdir='z', linestyle = 'dotted', zorder = -1)

ax.text(-0.1, 0, 0.1, "O", "x", size = 16)
ax.text(0.9, 0, 0.1, "P", "x", size = 16)
#ax.legend()
ax.grid(False)
plt.show()
fig.savefig('VelocityComponents.png')
```
