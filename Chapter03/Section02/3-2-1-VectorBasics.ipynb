## Plotting a vector in 2 D

We want to represent a vector in the Cartesian plane.
We assume that the components of the vector $\vec{A}$ are
$$ A_x, A_y$$

The code for producing the plot is based on the **quiver()** function of the **matplotlib** module.

```python
import matplotlib.pyplot as plt

Ax = 2
Ay = 1

fig, axs = plt.subplots()
axs.axis('equal')
axs.set(xlim=(-4, 4), ylim=(-4, 4))
axs.axhline()
axs.axvline()
axs.grid(which='major', axis = 'both')
plt.quiver(Ax, Ay, angles = 'xy',
           scale_units = 'xy', scale = 1,
           color = 'r')

plt.show()
```

### Applied vectors

Using the **quiver()** function is possible to assign the coordinates of the tail of the vector.

```python
import matplotlib.pyplot as plt

Ax = 2
Ay = 1

fig, axs = plt.subplots()
axs.axis('equal')
axs.set(xlim=(-4, 4), ylim=(-4, 4))
axs.axhline()
axs.axvline()
axs.grid(which='major', axis = 'both')

plt.quiver(Ax, Ay, angles = 'xy',
           scale_units = 'xy', scale = 1,
           color = 'r')
plt.quiver(1, 1, Ax, Ay, angles = 'xy',
           scale_units = 'xy', scale = 1,
           color = 'cyan')
plt.quiver(1, -1, Ax, Ay, angles = 'xy',
           scale_units = 'xy', scale = 1,
           color = 'C1')
plt.quiver(-3, 1, Ax, Ay, angles = 'xy',
           scale_units = 'xy', scale = 1,
           color = 'C2')
plt.quiver(-2, -2, Ax, Ay, angles = 'xy',
           scale_units = 'xy', scale = 1,
           color = 'm')
plt.show()

```

> It should be underlined that in the code reported above we have used different ways to specify the color of the arrow which represents the vector

* All the arrows represent the same vector, because they have the same length and the same direction.

* However, if we want to represent the force acting on a body, and in particular on a point-like body, we assume that the force is applied in a specific point.

### Vector addition

We can use the **numpy** module to perform operations between vectors.
Here we consider first the case of the sum of two vectors

$$ \vec{A} \equiv \left( A_x, A_y \right) $$

$$ \vec{B} \equiv \left( B_x, B_y \right) $$

If $\vec{C} = \vec{A} + \vec{B} $, then 

$$ C_x = A_x + B_x $$ 
$$ C_y = A_y + B_y $$ 

```python
import numpy as np

Ax = 1
Ay = 2
A = np.array([Ax, Ay])

Bx = 3
By = 1
B = np.array([Bx, By])

C = A + B

print()
print(A, '+', B, ' = ', C)
print()
```

#### The vector or its components?

* It can be useful to consider that we can use **the array** which represents the vector

```python
print(A)
```

* or we can use the components

```python
print(*A)
```

&nbsp;

### Plotting the sum of two vectors

Now, with the **quiver()** function, giving the components of the vectors, we can plot the vectors and their sum.

```python
import matplotlib.pyplot as plt


fig, axs = plt.subplots()

axs.set(xlim=(-5, 5), ylim=(-5, 5))
axs.axhline()
axs.axvline()
axs.grid(which='major', axis='both')
plt.quiver(*A, angles='xy', scale_units = 'xy', scale = 1, color = 'blue')
plt.quiver(*A, *B, angles='xy', scale_units = 'xy', scale = 1, color = 'red')
plt.quiver(*C, angles='xy', scale_units = 'xy', scale = 1, color = 'yellow' )
axs.set_aspect('equal')
plt.show()
```

This plot highlights the **tail to head rule** to sum vectors.

We place the tail of the second vector at the head of the first vector. The sum of the two vectors is the vector whose tail coincides with the tail of the first vector and whose head coincides with the head of the second vector.

### Multiplication by a scalar

If we multiply the vector $\vec{A}$ by a scalar $s$, we obtain a vector

$$ \vec{D} \equiv \left( s A_x, s A_y \right)$$

In particular, if we choose $s = -1$, we obtain **the opposite** of vector $\vec{A}$.

```python
D = -1 * A

print()
print(A)
print()
print(D)
print()
```

#### Graphical representation

```python
fig, axs = plt.subplots()

axs.set(xlim=(-5, 5), ylim=(-5, 5))
axs.axhline()
axs.axvline()
axs.grid(which='major', axis='both')
plt.quiver(*A, angles='xy', scale_units = 'xy', scale = 1, color = 'blue')
plt.quiver(*D, angles='xy', scale_units = 'xy', scale = 1, color = 'magenta' )
axs.set_aspect('equal')
plt.show()
```

### Magnitude

The magnitude of a vector $\vec{A}$ is denoted as $|\vec{A}|$ and can be evaluated as 

$$ |\vec{A}| = \sqrt{A_x^2 + A_y^2} $$

#### In Python

```python
print(np.sqrt(np.sum(A**2)))
```

#### or

```python
print(np.linalg.norm(A))
```

&nbsp;

### Dot product
The dot product between two vectors $\vec{A}$ and $\vec{B}$ is the scalar quantity 

$$ \vec{A}\cdot \vec{B} = A_x B_x + A_y B_y $$

#### Using Python and the numpy module

```python
print(A, B)
print(np.dot(A, B))
```

#### A particular case

As a particular case,

$$ \vec{A}\cdot\vec{A}  = A_x A_x + A_y A_y = |\vec{A}|^2$$

Thus, 
> we can exploit this feature for calculating the **magnitude** of the vector

as follows

```python
magnitude = np.sqrt(np.dot(A,A))
print(magnitude)
```

### Vector product

The vector product (or **cross product**) of two vectors $\vec{A}$ and $\vec{B}$ is denoted as

$$\vec{A}\times\vec{B}$$

* We draw the two vectors $\vec{A}$ and $\vec{B}$ with their tails at the same point

* The two vectors lie in the same plane. 
* We consider the angle $\theta$ as the smaller of the two possible angles between $\vec{A}$ and $\vec{B}$

- The magnitude of the vector product is given by 

$$ A B \sin{\theta} $$

- The direction is perpendicular to the plane of the vectors $\vec{A}$ and $\vec{B}$ and it is such that $\vec{A}$ has to rotate counterclockwise by the angle $\theta$ to superimpose to $\vec{B}$.

&nbsp;

If the components of the two vectors $\vec{A}$ and $\vec{B}$ are known, the components of their cross product are respectively

$$ A_y B_z - A_z B_y  $$
$$ A_z B_x - A_x B_z $$
$$ A_x B_y - A_y B_x $$

The vector product can be also written in a compact form as the following determinant:

$$ \begin{vmatrix}
\hat{u}_x & \hat{u}_y & \hat{u}_z\\
A_x & A_y & A_z\\
B_x & B_y & B_z
  \end{vmatrix}
$$


### Plotting the vector product

We need to construct a 3D plot!

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(1, (12,8))
ax = fig.gca(projection='3d')

a = np.array([1, 0, 0])
b = np.array([0, 1, 0])
m = np.cross(a, b)

ax.quiver(0, 0, 0, *a, length=1, color = 'green', linewidth = 4)
ax.quiver(0, 0, 0, *b, length=1, color = 'blue', linewidth = 4)
ax.quiver(0,0,0, *m, length = 1, color = 'red', linewidth = 4)

ax.set_xlim3d(-2, 2)
ax.set_ylim3d(-2,2)
ax.set_zlim3d(-2, 2)

plt.show()
```

### The cross product is *anticommutative*

This means that
$$ \vec{B}\times\vec{A} = - \vec{A}\times\vec{B} $$

The order of the two vectors involved in the cross product matters

&nbsp;

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(1, (12,8))
ax = fig.gca(projection='3d')

a = np.array([1, 0, 0])
b = np.array([0, 1, 0])
m = np.cross(a, b)

L = np.cross(b, a)

ax.quiver(0, 0, 0, *a, length=1, color = 'green', linewidth = 4)
ax.quiver(0, 0, 0, *b, length=1, color = 'blue', linewidth = 4)
ax.quiver(0,0,0, *m, length = 1, color = 'red', linewidth = 4)

ax.quiver(0,0,0, *L, length = 1, color = 'orange', linewidth = 4)

ax.set_xlim3d(-2, 2)
ax.set_ylim3d(-2,2)
ax.set_zlim3d(-2, 2)

plt.show()
```
