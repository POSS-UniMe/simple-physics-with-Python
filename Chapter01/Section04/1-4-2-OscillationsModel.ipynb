# Oscillations

## <span style="color:blue"> Model </span>

### Describing the phenomenon
How to describe the position of a point-like body as a function of time in the case of an oscillatory motion which occurs along a straight line?

The equation giving the coordinate $x$ as a function of the time $t$ is of the kind

$$ x = A \cos(\omega t + \phi)$$


### Calculations with Python
We want to look in detail to what happens when the value of the variable $t$ is changed:
- we choose increasing values of $t$ starting from the initial instant $t = 0 $;
- we calculate the value of the $x$ coordinate;
- we show in a graph the position assumed by the point-like body in different time instants.

&nbsp;



```python
import numpy as np
import matplotlib.pyplot as plt

omega = 2*np.pi / 10   # angular frequency (rad/s)
A = 0.15        # amplitude (m)
phi = 0         # initial phase (rad)

t = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.7, 6.3, 7.1, 7.7, 8.2, 8.7])   # time (s)
x = A * np.cos(omega*t + phi)
y = np.zeros(len(t)) 

fig = plt.figure(1, (14, 3))
ax = fig.add_subplot(111)

plt.plot(x, y, 'o')
plt.axhline(color = 'gray')
plt.axvline(color = 'cyan')
plt.xlim(-0.2, 0.2)
ax.set_xlabel('x (m)', fontsize = 14)

for i, time in enumerate(t):
    #print(i, time)
    x_coord = A * np.cos(omega*time + phi)
    time = str(time)+' s'
    ax.annotate(time, xy = (x_coord, 0.01), xycoords = 'data', rotation = 90, fontsize = 11 )
```

In this plot we can read the value of the time, and follow the position of the body at increasing values of $t$.

Showing simultaneously the positions occupied in many different instants we will obtain a grah of the **trajectory**

&nbsp;

## The *x-t* curve

The graph of the values of the coordinate $x$, giving the position, as a function of time $t$ in the Cartesian plane $tOx$ is the *x-t* **curve**.

```python
T = 2 * np.pi / omega
t = np.linspace(0, 3*T, 101)
x = A * np.cos(omega* t)

plt.figure(2)
plt.plot(t, x, '-', color = 'blue', linewidth = 3)
plt.xlabel('time $t$ (s)', fontsize = 14)
plt.ylabel(' $x$ (m)', fontsize = 14)
plt.axhline(color = 'gray')
plt.axvline(color = 'gray')
plt.axvline(x = T)
plt.axvline(x = 2 * T)

print()
print ('period T = ', T, 's')
print()
```

### What happens on varying the **angular frequency** ?

We choose two different values of the angular frequency and we show simultaneously in the same plot the two corresponding $x-t$ curves 


```python
omega = 0.628
T = 2*np.pi/omega
t = np.linspace(0, 2 * T, 101)
x = A * np.cos(omega*t)

omega1 = 1.2


x_star = A * np.cos(omega1*t)

plt.figure(3)
plt.plot(t, x, '-', color = 'blue', linewidth = 3)
plt.plot(t, x_star, '--', color = 'orange', linewidth = 3)
plt.xlabel('time $t$ (s)')
plt.ylabel(' $x$ (m)')
plt.axhline(color = 'gray')
plt.axvline(color = 'gray')
```

#### Relations between the variables

Angular frequency  $\omega$

Period $$ T = \dfrac{2 \pi}{\omega} $$

Frequency $f$

$$ f = \dfrac{1}{T} $$

$$ \omega = 2 \pi f $$

&nbsp;

### What happens on varying the amplitude A ?

We compare in the same plot the $x-t$ curves obtained choosing two different values of the amplitude, while keeping fixed the angular frequency (and the initial phase).

```python
omega = 0.628
T = 2*np.pi/omega
t = np.linspace(0, 2 * T, 101)
x = A * np.cos(omega*t)
A1 = 0.10
x_star = A1 * np.cos(omega*t)

plt.figure(4)
plt.plot(t, x, '-', color = 'blue', linewidth = 3)
plt.plot(t, x_star, linestyle = 'dotted', color = 'green', linewidth = 3)
plt.axhline(color = 'gray')
plt.xlabel('time  $t$ (s)')
plt.ylabel(' $x$ (m)')
```

### What happens on varying the *initial phase*  ?

We compare the $x-t$ curves corresponding to two different values of the initial phase, for assigned values of the amplitude and of the angular frequency.

&nbsp;

```python
T = 2 * np.pi / omega
t = np.linspace(0, 2 * T, 101)
x = A * np.cos(omega*t)

phi = np.pi/2

x_star = A * np.cos(omega*t + phi)

#z = x + x_star

plt.figure(5)
plt.plot(t, x, '-', color = 'blue', linewidth = 3)
plt.plot(t,x_star, linestyle = 'dashdot', color = 'lime', linewidth = 3)
#plt.plot(t, z, '-', color = 'yellow', linewidth = 5)
plt.axhline(color = 'gray')
plt.xlabel('time  $t$ (s)')
plt.ylabel(' $x$ (m)')
plt.axvline(color = 'gray')
plt.axvline(x = T, color = 'cyan')
```

## Equations of motion

### Velocity

From the definition of the **instantaneous velocity** $ v = dx/dt $, it follows that

$$ v = \dfrac{d}{dt}\left[ A \cos(\omega t + \phi) \right] =  
       - A \omega \sin(\omega t + \phi) $$
       
So, also the velocity values oscillate between a maximum value $v_{max} = +A\omega$ and a minimum value $v_{min} = - A \omega$, with the same period which characterizes the variations of the coordinate $x$ with respect to time.

&nbsp;

```python
T = 2 * np.pi / omega
t = np.linspace(0, 2 * T, 101)
x = A * np.cos(omega*t)

v = - A*omega*np.sin(omega*t)

print('\nMaximum velocity v_max = ', A*omega, ' (m/s) \n')

plt.figure(6,(8, 8))
plt.subplot(2, 1, 1)
plt.plot(t, x, '-', color='blue', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.axvline(x = T, color = 'cyan', linestyle='dashdot')
plt.xlabel('time  $t$ (s)')
plt.ylabel(' $x$ (m)')

plt.subplot(2, 1, 2)
plt.plot(t, v, '-', color='red', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.axvline(x = T, color = 'cyan', linestyle='dashdot')
plt.xlabel('time  $t$ (s)')
plt.ylabel('velocity $v$ (m/s)')
```

### Acceleration
From the definition of the instantaneous acceleration $ a = dv/dt $, it follows that

$$ a = \dfrac{d}{dt}\left[ - A\omega \sin(\omega t + \phi) \right] =  
       - A \omega^2  \cos(\omega t + \phi) $$
       
The values of the acceleration oscillate between a maximum value  $a_{max} = A\omega^2$ and a minimum value $a_{min} = -A\omega^2$, with the same period $T$ that characterizes the behaviour of the velocity $v$ and of the coordinate $x$.

&nbsp;

```python
T = 2 * np.pi / omega
t = np.linspace(0, 2 * T, 101)
a = - A * omega*omega* np.cos(omega*t)

print('\nMaximum acceleration a_max = ', A*omega*omega, ' (m/s^2) \n')

plt.figure(7,(8, 8))

plt.subplot(3, 1, 1)
plt.plot(t, x, '-', color='blue', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.axvline(x = T, color = 'cyan', linestyle='dashdot')
plt.xlabel('time  $t$ (s)')
plt.ylabel(' $x$ (m)')

plt.subplot(3, 1, 2)
plt.plot(t, v, '-', color='red', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.axvline(x = T, color = 'cyan', linestyle='dashdot')
plt.xlabel('time  $t$ (s)')
plt.ylabel('velocity $v$ (m/s)')

plt.subplot(3, 1, 3)
plt.plot(t, a, '-', color='green', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.axvline(x = T, color = 'cyan', linestyle='dashdot')
plt.xlabel('time  $t$ (s)')
plt.ylabel(' $a$ (m/s$^2$)')
print()
```

&nbsp;

### Velocity as a function of the position - Acceleration as a function of the position

&nbsp;

```python
T = 2 * np.pi / omega
t = np.linspace(0, 2 * T, 201)
x = A * np.cos(omega*t)
v = - A*omega*np.sin(omega*t)
a = - A * omega*omega* np.cos(omega*t)

print()

plt.figure(8,(13, 8))

plt.subplot(2, 2, 1)
plt.plot(x, v, '-', color='red', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.xlabel(' $x$ (m)')
plt.ylabel('velocity $v$ (m/s)')

plt.subplot(2, 2, 3)
plt.plot(x, a, '-', color='green', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.xlabel('$x$ (m)')
plt.ylabel(' $a$ (m/s$^2$)')

plt.subplot(2, 2, 2)
plt.plot(x, np.abs(v), '-', color='red', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.xlabel(' $x$ (m)')
plt.ylabel('speed $|v|$ (m/s)')
plt.ylim(ymin=-0.105, ymax = 0.105)

plt.subplot(2, 2, 4)
plt.plot(x, np.abs(a), '-', color='green', linewidth = 3)
plt.axhline(color='gray')
plt.axvline(color='gray')
plt.xlabel('$x$ (m)')
plt.ylabel(' $|a|$ (m/s$^2$)')
plt.ylim(ymin=-0.065, ymax = 0.065)
print()
```
