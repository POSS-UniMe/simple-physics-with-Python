## Discussione del problema

### Calcolo dell'accelerazione
L'accelerazione √® definita come rapporto fra la variazione di velocit√† $\Delta v$ e la durata dell'intervallo di tempo $\Delta t$ in cui essa avviene.
    $$ a = \dfrac{\Delta v}{\Delta t} = \dfrac{v_f - v_i}{t_f - t_i} $$
Per descrivere il moto dell'aereo lungo la pista di atterraggio utilizziamo un asse $x$ parallelo alla pista ed orientato nel senso di marcia dell'aereo. Notiamo che la velocit√† finale √® nulla e la velocit√† iniziale √® positiva, quindi la variazione di velocit√† √® negativa e l'accelerazione sar√† negativa.
$$ a = \dfrac{-v_0}{\Delta t} $$


### Calcolo della distanza d'arresto
La distanza di arresto pu√≤ essere ottenuta tramite le leggi del moto uniformemente accelerato

$$ x = x_0 + v_0\cdot(t - t_0) + \dfrac{1}{2}a\cdot(t - t_0)^2 $$

Ricordiamo che i valori di $x_0, t_0$ e $v_0$ sono noti e che il valore dell'accelerazione $a$ √® stato calcolato nella prima parte del problema. Per calcolare il valore della coordinata $x$ in corrispondenza della quale l'aereo si ferma  baster√† sostituire il valore di $(t-t_0)$ che √® dato dal testo del problema.

La distanza percorsa fra il punto in cui l'aereo tocca la pista di atterraggio ed il punto in cui l'aereo di ferma √® la **distanza di arresto**. Si ha quindi 

$$ d = x - x_0 $$

cio√®

$$ d = v_0\cdot(t - t_0) + \dfrac{1}{2}a\cdot(t - t_0)^2 $$

&nbsp;

Anzich√® sostituire il valore calcolato di $a$ si pu√≤ utilizzare il calcolo simbolico. Considerando l'espressione dell'accelerazione $a$, segue

$$ d = v_0 \Delta t + \dfrac{1}{2}\dfrac{\Delta v}{\Delta t}(\Delta t)^2 $$

e poich√® in questo caso si ha $\Delta v = 0 - v_0$, si ottiene

$$ d = v_0 \Delta t - \dfrac{1}{2}v_0 \Delta t $$

quindi

$$ d = \dfrac{1}{2}v_0 \Delta t $$

### Metodo alternativo
Considerando la relazione

$$ v^2 = v_0^2 + 2 a\cdot(x - x_0) $$

in questo caso si ottiene

$$ x - x_0 = \dfrac{-v_0^2}{2 a} $$

Il valore dell'accelerazione $a$ √® noto dalla prima parte del problema e $v_0$ √® un dato noto del problema, quindi con questa equazione di pu√≤ calcolare la distanza di arresto $d$ che √® pari ad $(x-x_0)$.

Questa equazione evidenzia che

- la distanza di arresto √® direttamente proporzionale al quadrato della velocit√† iniziale

- la distanza di arresto √® inversamente proporzionale al modulo dell'accelerazione, che dipende dall'intensit√† dell'azione frenante

- la quantit√† $d = x - x_0$ √® positiva, in quanto il fattore $-v_0^2$ √® negativo e l'accelerazione $a$ √® negativa

&nbsp;

D'altra parte, utilizzando il calcolo simbolico si pu√≤ scrivere

$$ x - x_0 = -\dfrac{v_0^2}{2} \dfrac{\Delta t}{\Delta v} $$

ovvero

$$ x - x_0 = -\dfrac{v_0^2}{2} \dfrac{\Delta t}{-v_0} $$

da cui infine 

$$ x - x_0 = \dfrac{1}{2} v_0 \Delta t $$



## Calcoli con Python

```python
vi = 234     # initial speed (km/h)
v0 = vi/3.6  # initial speed (m/s)
Dt = 2       # time lapse (s)
x0 = 0

a = - v0 / Dt             # acceleration
x = x0 + 0.5 * v0 * Dt    # final position
d = x - x0                # distance

print('\ninitial speed v0 = ', v0, '(m/s)')
print('\nacceleration a = ', a, '(m/s^2)')
print('\ndistance d = ', d, '(m)', '\n')

```

# Approfondimento
### Rappresentazione grafica delle equazioni del moto

```python
import numpy as np
import matplotlib.pyplot as plt

t0 = 0

print()
print('acceleration = ', a, ' (m/s^2) \n')
numpoints = 100

t = np.linspace(t0,t0 + Dt,numpoints)

v = v0 + a * (t - t0)

x = x0 + v0 * (t - t0) + 0.5 * a * (t - t0)**2

fig1 = plt.figure(figsize=(7, 13))

plt.subplot(3, 1, 1)
plt.axhline(y = a, xmin = 0.05, xmax = 0.95, color = 'green', linewidth = 3)
plt.axhline(y=0, color = 'gray', zorder = -1)
plt.axvline(x=t0, zorder = -1, linestyle = '--')
plt.axvline(x=t0+Dt, zorder = -1, linestyle = '--')
plt.xlabel('time  $t$ (s)')
plt.ylabel('$a$ (m/s$^2$)')

plt.subplot(3, 1, 2)
plt.plot(t, v, '-', color = 'red', linewidth = 2)
plt.axhline(color = 'gray', zorder = -1)
plt.axvline(color = 'gray', zorder = -1)
plt.xlabel('time  $t$ (s)')
plt.ylabel('$v$ (m/s)')

plt.subplot(3, 1, 3)
plt.plot(t, x, '-', color = 'blue', linewidth = 2)
plt.axhline(color = 'gray', zorder = -1)
plt.axhline(y = d, xmin=0.05, xmax = 0.95, color = 'orange', linestyle = '--', zorder = -1)
plt.axvline(color = 'gray', zorder = -1)
plt.xlabel('time  $t$ (s)')
plt.ylabel('$x$ (m)')
```
